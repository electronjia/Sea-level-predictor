```python
# Using  classes and methods

import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

class sea_level_predictor:
    def __init__(self,directory):
        self.pwd = directory
        
    def draw_plot(self):
        # Read data from file
        self.df = pd.read_csv(self.pwd)
        self.x = self.df['Year']
        self.y = self.df['CSIRO Adjusted Sea Level']
        
        # Create scatter plot
        figure, axis  = plt.subplots()
        axis.plot(self.x, self.y, 'o', label='original')
    
        # Create first line of best fit
        slope, intercept,r, p, se = linregress(self.x, self.y)
        self.x1 = range(self.x.min(), 2050, 1 )
        self.y1 = intercept + slope * self.x1
        
        axis.plot(self.x1, self.y1, label=f'{self.x.min()}-2050 fit' )
    
        # Create second line of best fit
        self.x = self.df.loc[self.df['Year'] >= 2000, 'Year']
        self.y = self.df.loc[self.df['Year'] >= 2000, 'CSIRO Adjusted Sea Level']
        
        slope, intercept,r, p, se = linregress(self.x, self.y)
        self.y2 = intercept + slope * self.x1
        
        axis.plot(self.x1, self.y2, label=f'{self.x.min()}-2050 fit' )
    
        # Add labels and title
        axis.set(xlabel='Year', ylabel='Sea Level (inches)', title='Rise in Sea Level')
        axis.legend(loc='best')
   
        # Save plot and return data for testing (DO NOT MODIFY)
        # plt.savefig('sea_level_plot.png')
        return plt.gca()
```

```python
sea_level_predictor('epa-sea-level.csv').draw_plot()
```
